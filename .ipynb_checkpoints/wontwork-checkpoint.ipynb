```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
#import matplotlib.pyplot as plt # we only need pyplot
import matplotlib as mp # we only need pyplot
from sklearn.cluster import kmeans_plusplus
sb.set() # set the default Seaborn style for graphics
```

#  Using Box-Whisker to detect and remove outliers.

```python
def remove_boxliers(df, li):
    
    dup = df
    q_low = []
    q_high = []
    
    for i in li:
        # print(i)
        #print(df[i])
        q1 = dup[i].quantile(0.25)
        q3 = dup[i].quantile(0.75)
        iqr = q3 - q1
        low_lim = q1 - 1.5 * iqr
        upp_lim = q3 + 1.5 * iqr
        q_low.append(low_lim)
        q_high.append(upp_lim)
    
    #print(q_low)
    #print(q_high)
    n = 0
    for i in li:
        df = df[ ( df[i] > q_low[n]) & (df[i] < q_high[n]) ]
        n += 1
        
    return df
```

```python
df = pd.read_csv('cardio.csv')
df.info()
```

```python
rem = remove_boxliers(df, ['BMI', 'AP_LO', 'AP_HI'])
rem.info()
```

```python
numeric_DF = pd.DataFrame(rem[['Age', 'Height', 'Weight', 'BMI', 'AP_HI', 'AP_LO']])
numeric_DF.info()
print()
print(numeric_DF)
```

```python
f, axes = mp.pyplot.subplots(6, 3, figsize=(18, 24))

count = 0
for var in numeric_DF:
    sb.boxplot(data = numeric_DF[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = numeric_DF[var], ax = axes[count,1])
    sb.violinplot(data = numeric_DF[var], orient = "h", ax = axes[count,2])
    count += 1
```

# Using Pairplot to detect and remove multi-variate outliers.

```python
# Extract the Features from the Data
X = numeric_DF

# Plot the Raw Data on 2D grids
sb.pairplot(X)
```

```python
# Import LocalOutlierFactor from sklearn.neighbors
from sklearn.neighbors import LocalOutlierFactor

# Set the Parameters for Neighborhood
num_neighbors = 20      # Number of Neighbors
cont_fraction = 0.05    # Fraction of Anomalies

# Create Anomaly Detection Model using LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors = num_neighbors, contamination = cont_fraction)

# Fit the Model on the Data and Predict Anomalies
lof.fit(X)
```

```python
# Predict the Anomalies
labels = lof.fit_predict(X)

# Append Labels to the Data
X_labeled = X.copy()
X_labeled["Anomaly"] = pd.Categorical(labels)

# Summary of the Anomaly Labels
sb.countplot(x = X_labeled["Anomaly"])
```

```python
# Visualize the Anomalies in the Data
sb.pairplot(X_labeled, vars = X.columns.values, hue = "Anomaly")
```

```python
X_labeled.info()
```

```python
#drop anomaly from LOF
numeric_DF2 = X_labeled.drop(X_labeled[X_labeled.Anomaly == -1].index)
numeric_DF2 = numeric_DF2.drop(['Anomaly'], axis = 1)
numeric_DF2 = numeric_DF2.reset_index(drop=True)
numeric_DF2.info()
print()
print(numeric_DF2)
```

```python
f, axes = mp.pyplot.subplots(6, 3, figsize=(18, 24))

count = 0
for var in numeric_DF2:
    sb.boxplot(data = numeric_DF2[var], orient = 'h', ax = axes[count,0])
    sb.histplot(data = numeric_DF2[var], ax = axes[count,1])
    sb.violinplot(data = numeric_DF2[var], orient = "h", ax = axes[count,2])
    count += 1
```
